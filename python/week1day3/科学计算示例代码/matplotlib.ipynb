```python
import numpy as np

x = np.linspace(-100, 100, 10000)
y = x ** 2
plt.plot(x, y)
plt.show()
```

#### plt.plot(x, y, [fmt])

* 可选参数fmt：设置格式
    * 线型：- -- -. : 等 
    * 点型： v,^,s,*,H,+,x,D,o 等 
    * 颜色：b,g,r,y,k,w 等 
    * 也可以通过linestyle, marker, color参数分别传入

```python
import matplotlib.pyplot as plt

# 清除当前图像
plt.clf()
# plt.plot(x, y, '-+g') # 实线，+点，绿色
plt.plot(x, y, linestyle=':', marker='v', color='#2f3b5a') # 点虚线，倒三角点，自定义颜色
plt.show()
```

```python
# 绘制多条线：在show()前执行多次plot()

x = np.linspace(-100, 100, 10000)
plt.clf()
plt.plot(x, x ** 2, color='red')
plt.plot(x, x * 0, color='green')
plt.plot(x, 2 * x ** 2, color='black')
plt.show()
```

#### 图像标注：

* 设置图像标题：plt.title()
* 设置x轴名称：plt.xlabel()
* 设置y轴名称：plt.ylabel()
* 设置x轴范围：plt.xlim()
* 设置y轴范围：plt.ylim()
* 设置x轴刻度：plt.xticks()
* 设置y轴刻度：plt.yticks()
* 设置曲线图例：plt.legend()

```python
plt.clf()
x = np.linspace(-100, 100, 10000)
plt.plot(x, x ** 2, color='red', label='y=x^2')
plt.plot(x, x, color='green', label='y=x')
plt.plot(x, 10*np.sin(x), color='black', label='y=10sin(x)')

# 标题和坐标轴标签
plt.title("Function Images")
plt.xlabel('x')
plt.ylabel('y')

# 坐标轴范围
plt.xlim(-100, 100)
plt.ylim(-100, 100)

# 坐标轴刻度
plt.xticks(np.arange(-100, 101, 50))
plt.yticks(np.arange(-100, 101, 25))

# 配合plot函数的label参数
plt.legend() 

plt.show()
```

```python
# 饼图
import matplotlib.pyplot as plt
plt.pie([10,20,30,40], labels=['a','b','c','d'], 
autopct="%.2f%%", explode=[0.1,0,0.1,0])
plt.axis('equal')
plt.show()
```

```python
# 柱状图：plt.bar(x, height, width=0.8)

data = [32, 48, 21, 100]
labels = ['Jan','Feb','Mar','Apr']
plt.bar(np.arange(len(data)), data, width=0.4, align='center')
plt.xticks(np.arange(len(data)), labels)
plt.show()
```

```python
# 散点图：plt.scatter(x, y)

x = np.random.randn(100)
y = np.random.randn(100)
plt.scatter(x, y, marker='x', color='red')
plt.show()
```

```python
# 直方图：plt.hist(x, bins)

x = np.random.normal(100, 15, 100000)
n, bins, patches = plt.hist(x, 50, color='blue', edgecolor='black')
plt.show()
```
