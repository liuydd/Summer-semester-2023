```python
# 导入sklearn包里相关工具
from sklearn import datasets
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
# 加载数据集
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# 划分数据集
## test_size：将原始数据集切分为70%的训练集和30%的测试集
## stratify分层：使测试集和训练集里各个类别的数据比例与原始数据集里比例一致
## shuffle：是否提前打乱数据
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)
print('X_test: ', X_test, type(X_test))
print('y_test: ', y_test, type(y_test))
```

```python
# 数据预处理
## 导入需要的数据标准化工具包
from sklearn.preprocessing import StandardScaler
## 转换器实例
scaler = StandardScaler()
## 拟合和转化
new_X_train = scaler.fit_transform(X_train)
new_X_test = scaler.fit_transform(X_test)
print('new_X_test: ', new_X_test, type(new_X_test))
```

```python
# 数据预处理
## 数据归一化
scaled_X = preprocessing.scale(X_train) ## 数据缩放
normalized_X = preprocessing.normalize(X_train) ## 数据归一化
standardized_X = preprocessing.scale(X_train) ## 数据标准化
print('scaled_X: ', scaled_X)
print('normalized_X: ', normalized_X)
print('standardized_X: ', standardized_X)
```

```python
# 监督学习算法
## 回归
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
print('lr: ', lr)
lr.fit(X_train, y_train)
y_pred_test = lr.predict(X_test)
print('y_pred_test: ', y_pred_test)
```

```python
# 监督学习算法
## 逻辑回归
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
print('lr: ', lr)
lr.fit(X_train, y_train)
y_pred_test = lr.predict(X_test)
print('y_pred_test: ', y_pred_test)
print('result: ', metrics.classification_report(y_test, y_pred_test))
print('confusion matrix: ', metrics.confusion_matrix(y_test, y_pred_test))
```

```python
## Taken from https://scikit-learn.org/stable/auto_examples/cluster/plot_bisect_kmeans.html#sphx-glr-auto-examples-cluster-plot-bisect-kmeans-py

import matplotlib.pyplot as plt

from sklearn.cluster import BisectingKMeans, KMeans
from sklearn.datasets import make_blobs

print(__doc__)


# Generate sample data
n_samples = 10000
random_state = 0

X, _ = make_blobs(n_samples=n_samples, centers=2, random_state=random_state)

# Number of cluster centers for KMeans and BisectingKMeans
n_clusters_list = [4, 8, 16]

# Algorithms to compare
clustering_algorithms = {
    "Bisecting K-Means": BisectingKMeans,
    "K-Means": KMeans,
}

# Make subplots for each variant
fig, axs = plt.subplots(
    len(clustering_algorithms), len(n_clusters_list), figsize=(12, 5)
)

axs = axs.T

for i, (algorithm_name, Algorithm) in enumerate(clustering_algorithms.items()):
    for j, n_clusters in enumerate(n_clusters_list):
        algo = Algorithm(n_clusters=n_clusters, random_state=random_state, n_init=3)
        algo.fit(X)
        centers = algo.cluster_centers_

        axs[j, i].scatter(X[:, 0], X[:, 1], s=10, c=algo.labels_)
        axs[j, i].scatter(centers[:, 0], centers[:, 1], c="r", s=20)

        axs[j, i].set_title(f"{algorithm_name} : {n_clusters} clusters")


# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
```

```python
# 超参数调优
from sklearn.model_selection import GridSearchCV
from sklearn import svm
svc = svm.SVC()
params = {'kernel': ['linear', 'rbf'], 'C': [1, 10]}
grid_search = GridSearchCV(svc, params, cv=5)
grid_search.fit(X_train, y_train)
grid_search.best_params_
```

```python
# 随机搜索
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
svc = svm.SVC()
param_dist = {'kernel': ['linear', 'rbf'], 'C': randint(1, 20)}
random_search = RandomizedSearchCV(svc, param_dist, n_iter=10)
random_search.fit(X_train, y_train)
random_search.best_params_
```
